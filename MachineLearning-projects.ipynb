```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np


%matplotlib inline

print("Libraries imported successfully!")
```

```python
# Load the data
cars = pd.read_csv('mtcars.csv')

# Display first few rows
cars.head()
```

## Data Overview
The dataset contains specifications for 32 different car models including:
- MPG (Miles per Gallon)
- Number of cylinders
- Horsepower
- Weight
- Transmission type

```python
# Rename columns if needed
cars.columns = ['car_name','mpg','cyl','disp','hp','drat','wt','qsec','vs','am','gear','carb']

# Check basic info
cars.info()

# Check for missing values
print("Missing values:")

print(cars.isnull().sum())
```

```python
# Create a bar plot of MPG
plt.figure(figsize=(12, 6))
cars['mpg'].plot(kind='bar')
plt.title('Miles Per Gallon by Car Model')
plt.xlabel('Car Index')
plt.ylabel('MPG')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
