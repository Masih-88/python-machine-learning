```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np


%matplotlib inline

print("Libraries imported successfully!")
```

```python
# Load the data
cars = pd.read_csv('mtcars.csv')

# Display first few rows
cars.head()
```

## Data Overview
The dataset contains specifications for 32 different car models including:
- MPG (Miles per Gallon)
- Number of cylinders
- Horsepower
- Weight
- Transmission type

```python
# Rename columns if needed
cars.columns = ['car_name','mpg','cyl','disp','hp','drat','wt','qsec','vs','am','gear','carb']

# Check basic info
cars.info()

# Check for missing values
print("Missing values:")

print(cars.isnull().sum())
```

```python
# Create a bar plot of MPG
plt.figure(figsize=(12, 6))
cars['mpg'].plot(kind='bar')
plt.title('Miles Per Gallon by Car Model')
plt.xlabel('Car Index')
plt.ylabel('MPG')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

cars = pd.read_csv('mtcars.csv')

# Display first few rows
cars.head()
cars.columns = ['car_name','mpg','cyl','disp','hp','drat','wt','qsec','vs','am','gear','carb']


mpg = cars['mpg']
mpg.plot()
plt.show()
```

```python
df = cars[['cyl', 'wt', 'mpg']]
df.plot()
plt.show()
```

```python
x = range(1, 10)
y = [1, 2, 3, 4, 0,4,3,2,1]
plt.bar(x, y)
plt.show()
```

```python
mpg = cars['mpg']
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
mpg.plot()



ax.set_xticklabels(cars['car_name'], rotation=60)
ax.set_xticks(range(32))

ax.set_xlabel('Car Names')
ax.set_ylabel('miles/gal')

ax.set_title('Miles Per Gallon by Car Model')
plt.show()
```

```python

cars.index = cars.car_name
mpg = cars['mpg']
mpg.plot(kind = 'hist')
plt.show()
```

```python
cars.plot(kind = 'scatter', x = 'hp', y = 'mpg', c= ['darkgray'], s=150)
plt.show()
```

```python
sns.regplot(x = 'hp', y = 'mpg', data = cars , scatter = True)
plt.show()
```

```python
sns.pairplot(cars)
plt.show()
```

```python
cars_subset = cars[['mpg', 'disp', 'hp', 'wt']]
sns.pairplot(cars_subset)
plt.show()
```
